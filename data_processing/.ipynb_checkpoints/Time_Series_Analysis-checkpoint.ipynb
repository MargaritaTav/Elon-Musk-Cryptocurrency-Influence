```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
all_tweets= pd.read_csv("musktweets.csv", error_bad_lines=False, index_col = [0])
```

```python
all_tweets.head()
```

```python
# Filter the data using keywords
all_tweets['MESSAGE'].fillna('', inplace=True)
keywords = ['Bitcoin', 'BTC', 'Crypto', 'Cryptocurrency', 'Dogecoin', 'Bit coin', 'Cryptomarket']

filtered_dataset = all_tweets[all_tweets['MESSAGE'].str.contains('|'.join(keywords), case=False)]
```

```python
# Upload Bitcoin dataset with historical data
btc_dataset = pd.read_csv("BTC-Hourly.csv")
```

```python
btc_dataset.drop('high', axis=1, inplace=True)
btc_dataset.drop('low', axis=1, inplace=True)
btc_dataset.drop('symbol', axis=1, inplace=True)
btc_dataset.drop('Volume BTC', axis=1, inplace=True)
btc_dataset.drop('Volume USD', axis=1, inplace=True)
```

```python
# Make sure that the timezone is the same
filtered_dataset.head(500)
btc_dataset['date'] = pd.to_datetime(btc_dataset['date'], utc = True)
all_tweets['date'] = pd.to_datetime(all_tweets['date'], utc = True)
all_tweets['date'] = all_tweets['date'].dt.floor('H')
```

```python
btc_dataset.dtypes
```

```python
# Upload Bitcoin dataset with more recent data
btc_dataset_new= pd.read_csv("bitcoin_new.csv")
```

```python
# Make sure that the new dataset has the same format 
btc_dataset_new['open'] = btc_dataset_new['open']/1000
btc_dataset_new['close'] = btc_dataset_new['close']/1000
```

```python
# Estimate the opening and closing prices hourly
btc_dataset_new['date'] = pd.to_datetime(btc_dataset_new['date'], utc = True)

# Set the timestamp as the index
btc_dataset_new.set_index('date', inplace=True)

# Resample the data to hourly intervals
btc_dataset_new = btc_dataset_new.resample('H').agg({'open': 'first', 'close': 'last'})

# Reset the index
btc_dataset_new.reset_index(inplace=True)

print(btc_dataset_new)
```

```python
btc_dataset_new.tail()
```

```python
# Concatenate two datasets
btc_dataset_final = pd.concat([btc_dataset_new, btc_dataset])
```

```python
btc_dataset_final = btc_dataset_final.drop_duplicates(subset = ['date'])
btc_dataset_final = btc_dataset_final.sort_values(by = ['date'], ascending = True)
```

```python
btc_dataset_final.head()
```

```python
btc_dataset_final = btc_dataset_final.reset_index(drop=True)
```

```python
btc_dataset_final.head()
```

```python
all_tweets = all_tweets.reset_index()
all_tweets.head()
```

```python
# Create a column to indicate whether a tweet was posted in the last hour
all_tweets['tweeted'] = all_tweets['date'].apply(
    lambda x: 1 if any((x - pd.Timedelta(hours=1) < btc_dataset_final['date']) & (btc_dataset_final['date'] <= x)) else 0
)

# Calculate the mean before and after each tweet
tweet_impact = []
for i, row in all_tweets.iterrows():
    if row['tweeted'] == 1:
        tweet_time = row['date']
        before_tweet = btc_dataset_final[btc_dataset_final['date'] < tweet_time].tail(12)
        after_tweet = btc_dataset_final[btc_dataset_final['date'] > tweet_time].head(12)
        mean_before_tweet = before_tweet['open'].mean()
        mean_after_tweet = after_tweet['close'].mean()
        tweet_impact.append((mean_before_tweet, mean_after_tweet))


for impact in tweet_impact:
    print(f"Mean before tweet: {impact[0]}, Mean after tweet: {impact[1]}")
```

```python
# Create a column to indicate whether a tweet was posted in the last hour
all_tweets['tweeted'] = all_tweets.index.isin(btc_dataset_final.index - pd.Timedelta(hours=1)).astype(int)

# Calculate the mean before and after each tweet
tweet_impact = []
for i, row in all_tweets.iterrows():
    if row['tweeted'] == 1:
        tweet_time = i
        before_tweet = btc_dataset_final[btc_dataset_final.index < tweet_time].tail(12)
        after_tweet = btc_dataset_final[btc_dataset_final.index > tweet_time].head(12)
        mean_before_tweet = before_tweet['open'].mean()
        mean_after_tweet = after_tweet['close'].mean()
        tweet_impact.append((i, mean_before_tweet, mean_after_tweet))

# Sort the tweet impact by the difference in means
tweet_impact.sort(key=lambda x: abs(x[2] - x[1]), reverse=True)

# Get the 15 most important tweets
top_tweets = tweet_impact[:15]

# Prepare data for visualization
dates = [tweet[0].strftime("%Y-%m-%d %H:%M:%S") for tweet in top_tweets]
mean_before = [tweet[1] for tweet in top_tweets]
mean_after = [tweet[2] for tweet in top_tweets]

# Create the line plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(dates, mean_before, marker='o', label='Mean Before')
ax.plot(dates, mean_after, marker='o', label='Mean After')
ax.set_xlabel('Tweet Timestamp')
ax.set_ylabel('Bitcoin Opening Price Mean')
ax.set_title('Impact of Elon Musk Tweets on Bitcoin Prices')
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
